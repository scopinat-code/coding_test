```python
#requsets를 사용할게요
import requests
# bs4에서 BeautyfulSoup만 사용할게요
from bs4 import BeautifulSoup 

#이 웹사이트에 모든정보를 requests(요청)해서 response(응답)을 받아주세요.
response = requests.get("https://startcoding.pythonanywhere.com/basic") 
#웹사이트의 html정보를 html변수에 저장해주세요
html = response.text
#html변수를 'html.parser'를 이용해서 나눈다음에 soup에 저장해주세요
soup = BeautifulSoup(html, "html.parser")
soup.select('.brand-name')
```
